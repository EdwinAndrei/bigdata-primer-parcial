# Ejercicio 1: Clasificación de edad.

# Enunciado: Determine la mayoria de edad sugun su edad.

# Lógica: se evaluara la edad usando condiciones if, elif, y else para determinar la edad.

```python
def clasificar_edad(edad):

    if edad < 13:
        return "Niño"

    elif edad <= 17:
        return "Adolescente"

    elif edad <= 59:
        return "Adulto"

    else:
        return "Adulto mayor"


edad = 21

categoria = clasificar_edad(edad)

print(f"Edad: {edad} años")
print(f"Categoría: {categoria}")
```

# Ejercicio 2: Conversor de temperatura.

# Enunciado: Convierta una temperatura de grados Celsius a Fahrenheit.

# Lógica: se utiliza la formula F = (C × 9/5) + 32 para que la temperatura se guarde en una variable y posterior mente se calcula el resultado. 

```python
def convertir_fahrenheit(celsius):

    fahrenheit = (celsius * 9/5) + 32

    return fahrenheit


temperatura_celsius = 40

resultado = convertir_fahrenheit(temperatura_celsius)

print(f"Temperatura en Celsius: {temperatura_celsius}°C")
print(f"Temperatura en Fahrenheit: {resultado}°F")
```

# Ejercicio 3: Calculadora básica.
# Enunciado: Realizar operaciones matemáticas básicas entre dos números.
# Lógica: Se utilizan las condiciones para determinar la operación que se realizará, al mismo tiempo, se valida que no se pueda dividir entre 0.

```python
def calculadora(num1, num2, operacion):

    if operacion == "+":
        return num1 + num2

    elif operacion == "-":
        return num1 - num2

    elif operacion == "*":
        return num1 * num2

    elif operacion == "/":

        if num2 != 0:
            return num1 / num2

        else:
            return "Error: división entre cero"

    else:
        return "Operación no válida"


numero1 = 37
numero2 = 7
operacion = "*"

resultado = calculadora(numero1, numero2, operacion)

print(f"Número 1: {numero1}")
print(f"Número 2: {numero2}")
print(f"Operación: {operacion}")
print(f"Resultado: {resultado}")
```

# Ejercicio 4: Número par o impar

**Enunciado:** determinar si un número entero es par o impar.

**Lógica:** se utiliza el operador módulo `%`, el cual permite obtener el residuo de una división. Si un número dividido entre 2 deja residuo 0, entonces es par. Si deja residuo diferente de 0, entonces es impar.

```python
def verificar_par_impar(numero):
    if numero % 2 == 0:
        return "Par"
    else:
        return "Impar"


numero = 15
resultado = verificar_par_impar(numero)

print(f"Número: {numero}")
print(f"Resultado: El número es {resultado}")
```

# Ejercicio 5: Tabla de multiplicar

**Enunciado:** generar la tabla de multiplicar de un número entero del 1 al 12.

**Lógica:** se define un número entero y se utiliza un ciclo `for` para repetir el proceso desde el 1 hasta el 12. En cada repetición, el número se multiplica por el valor actual del ciclo y se muestra el resultado.

```python
def tabla_multiplicar(numero):
    for i in range(1, 13):
        resultado = numero * i
        print(f"{numero} x {i} = {resultado}")


numero = 9

print(f"Tabla de multiplicar del {numero}")
tabla_multiplicar(numero)
```

# Ejercicio 6: Contador de vocales

**Enunciado:** contar cuántas vocales contiene una frase.

**Lógica:** se define una frase y se convierte a minúsculas con `.lower()` para que el programa pueda contar vocales aunque estén escritas en mayúsculas o minúsculas. Luego se recorre cada letra de la frase con un ciclo `for` y se verifica si la letra pertenece al conjunto de vocales.


```python
def contar_vocales(frase):
    vocales = "aeiou"
    contador = 0

    for letra in frase.lower():
        if letra in vocales:
            contador += 1

    return contador


frase = "Este es un ejercicio para contar vocales."
cantidad_vocales = contar_vocales(frase)

print(f"Frase: {frase}")
print(f"Cantidad de vocales: {cantidad_vocales}")
```

## Ejercicio 8: Lista de números positivos y negativos

**Enunciado:** clasificar 10 números contando cuántos son positivos, negativos y ceros.

**Lógica:** se recorre una lista con un `for` y se evalúa cada número con `if/elif/else` para contar cada categoría.


```python
numeros = [4, -2, 0, 7, -5, 0, 3, -1, 8, -9]

positivos = 0
negativos = 0
ceros = 0

for n in numeros:
    if n > 0:
        positivos += 1
    elif n < 0:
        negativos += 1
    else:
        ceros += 1

print(f"Positivos: {positivos}")
print(f"Negativos: {negativos}")
print(f"Ceros: {ceros}")
```

## Ejercicio 9: Mayor y menor de una lista

**Enunciado:** encontrar el número mayor y menor de una lista y calcular la diferencia entre ambos.

**Lógica:** se usan las funciones `max()` y `min()` que recorren la lista automáticamente y devuelven el valor más alto y más bajo respectivamente.


```python
numeros = [15, 3, 42, 8, 27, 5, 99, 11]

mayor = max(numeros)
menor = min(numeros)
diferencia = mayor - menor

print(f"Mayor: {mayor}")
print(f"Menor: {menor}")
print(f"Diferencia: {diferencia}")
```

### 11. Funcion para calcular descuento
El codigo podra funcionar con una funcion que sea capaz de procesar la informacion y devolver un valor de descuento.

En la funcion entra el dato del precio original junto con el porcentaje del descuento a aplicar. Se hace el calculo del precio original multiplicado por el porcentaje de descuento y este mismo se guarda en la clave **descuento** y luego se hace una resta del precio original menos el precio de descuento y este retorna el precio final.


```python
def calcular_descuento(precio_original, porcentaje_descuento):
    descuento = precio_original * (porcentaje_descuento / 100)
    precio = precio_original - descuento
    return precio

precio_final = calcular_descuento(100, 20)

print("Precio final:", precio_final)
```

### 12. Validacion de Contraseña
El codigo tendra la capacidad de validar una contraseña que sea ingresada por un usuario.

El programa valida si la contraseña ingresada cuenta con al menos **8 caracteres**, que contenga al menos **1 numero**, al menos **1 Mayuscula** y **1 Minuscula**. Si se cumplen estas condiciones, el programa indica con un mensaje si la contraseña es valida o no lo es. 

```python
contraseña = "BigData123"

cantidad = len(contraseña) >= 8
tiene_numero = any(caracter.isdigit() for caracter in contraseña)
tiene_mayuscula = any(caracter.isupper() for caracter in contraseña)
tiene_minuscula = any(caracter.islower() for caracter in contraseña)

if cantidad and tiene_numero and tiene_mayuscula and tiene_minuscula:
    print("La contraseña es válida")
else:
    print("La contraseña NO es válida")
```

# Ejercicio 13: Sistema simple de inventario

# Lógica: 
#Crear una lista donde cada elemento es un diccionario con las claves: código, nombre, precio, cantidad. Recorrer la lista con un bucle `for` para mostrar cada producto.

```python
productos = [
    {"codigo": 1, "nombre": "Lápiz", "precio": 0.5, "cantidad": 100},
    {"codigo": 2, "nombre": "Cuaderno", "precio": 2.0, "cantidad": 50},
    {"codigo": 3, "nombre": "Goma", "precio": 0.3, "cantidad": 80},
]

# Mostrar todos
print("Inventario:")
for p in productos:
    print(f"Código:{p['codigo']} - {p['nombre']} - ${p['precio']} - {p['cantidad']} uds")

codigo_buscar = 2
encontrado = None
for p in productos:
    if p["codigo"] == codigo_buscar:
        encontrado = p
        break

if encontrado:
    print(f"\nProducto encontrado: {encontrado['nombre']}, Precio: ${encontrado['precio']}, Stock: {encontrado['cantidad']}")
else:
    print("\nNo existe ese código")

total = sum(p["precio"] * p["cantidad"] for p in productos)
print(f"\nValor total del inventario: ${total}")
```

# Ejercicio 14: Registro de ventas

# Lógica:  
#Se usará una lista de diccionarios para almacenar las ventas.

```python
ventas = [
    {"cliente": "María", "producto": "Camisa", "cantidad": 2, "precio_unitario": 15},
    {"cliente": "Juan", "producto": "Pantalón", "cantidad": 1, "precio_unitario": 30},
    {"cliente": "María", "producto": "Zapatos", "cantidad": 1, "precio_unitario": 45},
    {"cliente": "Ana", "producto": "Bufanda", "cantidad": 3, "precio_unitario": 8},
]

for venta in ventas:
    venta["total"] = venta["cantidad"] * venta["precio_unitario"]

total_general = sum(v["total"] for v in ventas)

venta_max = max(ventas, key=lambda v: v["total"])
cliente_top = venta_max["cliente"]
monto_top = venta_max["total"]

print("Ventas detalladas:")
for v in ventas:
    print(f"{v['cliente']} compró {v['cantidad']} {v['producto']}(s) - Total: ${v['total']}")
print(f"\nTotal general vendido: ${total_general}")
print(f"Cliente con la compra más alta: {cliente_top} (${monto_top})")
```

# Ejercicio 15: Limpieza básica de texto


# Lógica:  
# Usar `strip()` para quitar espacios de los extremos.  
 #Usar `lower()` para pasar todo a minúsculas. 
 #Usar `split()` para dividir la frase en una lista de palabras.  
# Contar los elementos de esa lista con `len()`

```python
frase = "  Hola   Mundo  esto  es  Python  "

frase = frase.strip()
frase = frase.lower()
while "  " in frase:
    frase = frase.replace("  ", " ")
palabras = frase.split()
print("Frase limpia:", frase)
print("Cantidad de palabras:", len(palabras))
```
